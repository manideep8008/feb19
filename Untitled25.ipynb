App routing in Flask is the process of mapping URL paths to specific view functions that handle HTTP requests. It allows us to define different URL paths and bind them to corresponding functions to execute when the user accesses those URLs. We use app routes to provide a clean and organized way to handle different URL patterns within our web application. It simplifies URL handling and keeps the code modular, making it easier to maintain and understand the application's structure.

The url_for() function in Flask is used for URL building. It generates a URL for the specified endpoint (view function) by taking its name as an argument. This function abstracts the URL construction and ensures that changes to the URL structure are automatically reflected throughout the application.

Example Python code demonstrating the url_for() function:

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/about')
def about():
    return 'About Page'

@app.route('/contact')
def contact():
    return 'Contact Page'

if __name__ == '__main__':
    with app.test_request_context():
        print(url_for('home'))
        print(url_for('about'))
        print(url_for('contact'))
```
